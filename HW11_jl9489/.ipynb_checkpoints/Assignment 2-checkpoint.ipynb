```python
cdb = pd.read_csv(os.getenv("PUIDATA") + '/CHDB_data_city_NJ v3_0.csv')
cdb.head()
```

```python
cdb['city_name'].nunique()
```

```python
cdb_2 = cdb[cdb['metric_name']=='Children in Poverty']

cdb_2.head()
```

```python
cdb3 = cdb_2[cdb_2['group_name']=='total population']
```

```python
cdb3.head()
```

```python
cdb3.shape
```

```python
fig, ax = plt.subplots(figsize=(10,10))

names = cdb3['city_name']
values = cdb3['est']

ax.barh(names, values, alpha=0.5)
ax.set_xlabel('% Estimate of Children in Poverty', fontsize=12)
ax.set_ylabel('City Names', fontsize=12)
ax.set_title('Chilhood Poverty rates in New Jersey', fontsize=14)

plt.tight_layout()
```

Fig 1: The above chart shows the estimated percentage of children in poverty for each city in New Jersey.  This is a percentage of the total population, inclusive of race and gender. Camden has the highest child poverty rate at over 50%, while Clifton represents the lowest, with a rate of just over 10%.
