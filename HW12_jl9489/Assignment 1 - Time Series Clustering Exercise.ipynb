## Merge Datasets

```python
cbs = nyczipsdf.merge(df_all, right_on='zip', left_index=True, how='left')
```

```python
cbs['zip'] = cbs.zip.astype(int)
```

```python
cbs = pd.DataFrame(cbs)
```

```python
cbs = cbs.set_index('zip')
```

```python
cbs.shape
```

```python
cbs.head()
```

```python
cleancbs = cbs.dropna()

cleancbs.shape
```

```python
vals = ((cleancbs.T - cleancbs.mean(1)) / cleancbs.std(1)).T
```

```python
print("The means are as follows: \n")
print(vals.mean(axis=1))
print("The shape of the mean vector is: ", vals.mean(axis=1).shape)
print("The standard deviations are as follows: \n")
print(vals.std(axis=1))
print("The shape of the standard deviation vector is: ", vals.std(axis=1).shape)
```

## K-means Clustering

```python
nc=5

km = KMeans(n_clusters=nc, random_state=444).fit(vals)
```

```python
km.labels_.shape
```

```python
cluster_means = km.cluster_centers_
cluster_means.shape
```

```python
vals['cluster'] = km.labels_
vals.head()
```

```python
# Plotting clusters

fig, axes = plt.subplots(3, 2, figsize=(15,8))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=0.8)
for c,ax in enumerate(fig.axes):
    for i in range(len(vals[vals.cluster == c])):
        ax.plot(range(21), vals[vals.cluster == c].iloc[i,0:21], alpha=0.3)
        ax.plot(range(21), cluster_means[c], 'blue')
        plt.setp(axes, xticks=range(21), xticklabels=range(21))
        ax.set_xlabel('Years Since 1994')
        ax.set_ylabel('Establishments (standardized)')
        ax.set_title('Cluster {}'.format(c))
fig.suptitle('K-Means Clustering Results', fontsize=20)
```

Fig 1: The above figure is the output of a K-means clustering algorithm, for 5 clusters.  The data represents economic activity in NYC by zip code.  In the above graphs, each line of a plot represents a zip code.  Cluster 0 has the densest plot, corresponding to the highest count of unique zip codes within that cluster.  The data is extremely noisy, but among several of the clusters (0,3,4) there sems to be the biggest change near years 6 and 7.

## Agglomerative Clustering

```python
agc = AgglomerativeClustering(n_clusters=5, linkage='average').fit(vals.iloc[:, :-1])
labelsag = agc.labels_

agc
```

```python
vals['aggClusters'] = labelsag
```

```python
centroidsag = vals.groupby(['aggClusters']).mean().iloc[:, :-1]
```

```python
fig, axes = plt.subplots(3, 2, figsize=(15,8))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=0.8)
for c,ax in enumerate(fig.axes):
    for i in range(len(vals[vals.aggClusters == c])):
        ax.plot(range(21), vals[vals.aggClusters == c].iloc[i,0:21], alpha=0.3)
        ax.plot(range(21), centroidsag.iloc[c, :], 'blue')
        plt.setp(axes, xticks=range(21), xticklabels=range(21))
        ax.set_xlabel('Years Since 1994')
        ax.set_ylabel('Establishments (standardized)')
        ax.set_title('Cluster {}'.format(c))
fig.suptitle('Agglomerative Clustering Results', fontsize=20)
```

Fig 3: The above plot shows an alternate clustering method: agglomerative clustering.  In this plot, the third cluster (Cluster 2) retains the majority of zip codes.  Significant changes seem to take place around years 6 and 7 for all clusters, as well as years 11, 12, 13.

```python
agg_clust = vals.reset_index()[['zip', 'aggClusters']]
agg_clust.shape
```

```python
geom_clust2 = pd.merge(nyczipsshp, agg_clust, on='zip', how='right')
geom_clust2.shape
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,7))
fig.suptitle('Comparison of Clustering Methods with Six Clusters', fontsize=20)

geom_clust.plot(column='cluster', cmap='viridis', 
                edgecolor='black', categorical = True, legend=True, ax=ax1)
ax1.set_title('K-Means Clustering')
geom_clust2.plot(column='aggClusters', cmap='viridis', 
                edgecolor='black', categorical = True, legend=True, ax=ax2)
ax2.set_title('Agglomerative Clustering')

ax1.axis('off')
ax2.axis('off')
ax1.axis('equal')
ax2.axis('equal')
```

Fig 4: The above map compares the K-means and Agglomerative clustering methods.  As can be seen, there is significant difference between the two methods, in terms of what level of clustering contains the majority of zip codes.  However, the majority of zip codes that were clustered together using K-means were also clustered together using Aggolmerative clustering.  For Brooklyn, Queens, Bronx, and upper Manhattan, there is similarity irrespective of the algorithm.  This would be worth exploring further.
