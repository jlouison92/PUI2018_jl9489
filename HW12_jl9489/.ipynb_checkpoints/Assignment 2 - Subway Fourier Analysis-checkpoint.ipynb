Fig 3: The above figure shows the power spectrum of the Fourier transform of the dataset.  High spikes indicate potential periodicity in the data. In this case, the periodicity occurs approximately every 4 weeks, as well as at a much higher interval.  We can potentially surmise that in the aggregate, MTA ridership has seasonal trends that affect total counts on a monthly basis.

```python
fig = pl.figure(figsize=(15,5))
ax = fig.add_subplot(111)
for i in range(0,599):
    f = np.abs(np.fft.rfft(mta_weeklystations[i]))**2
    ax.plot(np.fft.rfftfreq(len(mta_weeklystations[i]), 1.0)[1:], (f[1:]), '-')
    ax.set_xticklabels(["%.1f"%(1/f) for f in ax.get_xticks()], rotation=90)
    pl.xlim(xmin=0, xmax=.05)
    pl.title("Stations Periodicity")
    pl.xlabel("weeks (1/f)")
    pl.ylabel("power");
```

Fig 4: Above is a more "zoomed-in" perspective of the biggest spike as seen in Figure 3.  Here we are able to further isolate the location of periodicity, as well as have a better understanding of how many stations contribute the the spike.  Although it is not extremely clear, it can be safely said that a handful of stations (<10 stations) contribute to the spike at this point.  The spike occurs at around 50 weeks, which would line up with an annual periodicity.
